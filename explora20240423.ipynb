# Sorting

```python
countries_df.sort_index(ascending=True)
```

```python
countries_df.sort_values('Population',ascending=False)
```

```python
columns=['Country','Population','Area (sq. mi.)']
countries_df.sort_values('Population',ascending=False).filter(items=columns)
```

# Grouping

```python
regions= countries_df.groupby('Region')
```

```python
(regions[['Population','Area (sq. mi.)']].sum()/1_000_000).round(1)
```

```python
regions['Pop. Density (per sq. mi.)'].mean().sort_values(ascending=False)
```

```python
regions.agg({'Population':np.sum,
             'GDP ($ per capita)':np.mean}).round(1).sort_values('GDP ($ per capita)',ascending=False)
```
