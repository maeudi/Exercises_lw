<h1> <strong>API del Mercado de Valores</strong> </h1>

<h3> 1) Importar Librerias </h3>

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib
```

<h3> 2) Importar API </h3>
<h4>La acción de Apple en los último 3 meses </h4>

```python
import requests

url = "https://cloud.iexapis.com/stable/stock/aapl/chart/3m?token=pk_8d403bca938d46e99fbd799b343f5ca4"
api_data = requests.get(url).json()
```

<h3> 3) Crear dataframe </h3>

```python
apple_stock_df = pd.DataFrame.from_dict(api_data)
apple_stock_df
```

<h3> 4) Conversión de fecha a objeto datetime </h3>

```python
apple_stock_df['date'] = pd.to_datetime(apple_stock_df['date'], format="%Y-%m-%d")
```

<h3> 5) Configuración de la columna de fecha como el index </h3>

```python
apple_stock_df = apple_stock_df.set_index('date')
```

<h3> 6) Graficos</h3>
<h4> Solo columnas "close"</h4>

```python
apple_stock_df['close'].plot()
```

<h4>  Graficar los valores de ‘open’, ‘close’, ‘high’, ‘low’ </h4>

```python
apple_stock_df[['open', 'close', 'high', 'low']].plot()
```

<h4> argumento figsize </h4>

```python
apple_stock_df[['open', 'close', 'high', 'low']].plot(figsize=(12,4))
```
